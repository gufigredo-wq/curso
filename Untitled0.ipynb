```python
def cifra_de_cesar(texto, deslocamento, modo='c'):
    resultado = ""
    # Se for para decifrar, basta inverter o deslocamento
    if modo.lower() == 'd':
        deslocamento = -deslocamento

    for char in texto:
        if char.isalpha():
            # Define se a base é 'A' (65) ou 'a' (97)
            base = ord('A') if char.isupper() else ord('a')
            # Calcula a nova posição com o módulo 26
            indice = (ord(char) - base + deslocamento) % 26
            resultado += chr(indice + base)
        else:
            # Mantém espaços, números e símbolos
            resultado += char

    return resultado

# --- INTERFACE DE ENTRADA ---
print("--- CIFRA DE CÉSAR INTERATIVA ---")
mensagem = input("Digite a palavra ou frase: ")
chave = int(input("Digite o deslocamento (número): "))
operacao = input("Deseja (C)ifrar ou (D)ecifrar? ").strip().lower()

if operacao in ['c', 'd']:
    final = cifra_de_cesar(mensagem, chave, operacao)
    print(f"\nResultado: {final}")
else:
    print("Operação inválida! Use 'C' para cifrar ou 'D' para decifrar.")
```

```python
def cifra_de_vigenere(texto, chave, modo='c'):
    resultado = ""
    chave = chave.upper()
    len_chave = len(chave)
    indice_chave = 0

    for char in texto:
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')

            # Determina o deslocamento com base na letra da chave
            deslocamento = ord(chave[indice_chave % len_chave]) - ord('A')

            if modo.lower() == 'd':
                deslocamento = -deslocamento

            # Calcula a nova posição com o módulo 26
            indice = (ord(char) - base + deslocamento) % 26
            resultado += chr(indice + base)
            indice_chave += 1 # Avança para a próxima letra da chave apenas se for um caractere alfabético
        else:
            # Mantém espaços, números e símbolos
            resultado += char

    return resultado

# --- INTERFACE DE ENTRADA ---
print("--- CIFRA DE VIGENÈRE INTERATIVA ---")
mensagem = input("Digite a palavra ou frase: ")
chave_vigenere = input("Digite a palavra-chave (somente letras): ")
operacao = input("Deseja (C)ifrar ou (D)ecifrar? ").strip().lower()

if operacao in ['c', 'd']:
    final_vigenere = cifra_de_vigenere(mensagem, chave_vigenere, operacao)
    print(f"\nResultado: {final_vigenere}")
else:
    print("Operação inválida! Use 'C' para cifrar ou 'D' para decifrar.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_first_degree_function():
    print("--- PLOTADOR DE FUNÇÃO DE PRIMEIRO GRAU (y = ax + b) ---")

    try:
        a = float(input("Digite o valor de 'a' (coeficiente angular): "))
        b = float(input("Digite o valor de 'b' (coeficiente linear): "))
    except ValueError:
        print("Entrada inválida! Por favor, digite números para 'a' e 'b'.")
        return

    # Gera pontos para o eixo x
    x = np.linspace(-10, 10, 400)  # Gera 400 pontos entre -10 e 10

    # Calcula os valores correspondentes de y
    y = a * x + b

    # Cria o gráfico
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'y = {a}x + {b}', color='blue')

    # Adiciona detalhes ao gráfico
    plt.title('Gráfico da Função de Primeiro Grau')
    plt.xlabel('Eixo X')
    plt.ylabel('Eixo Y')
    plt.grid(True)
    plt.axhline(0, color='black', linewidth=0.5) # Eixo X
    plt.axvline(0, color='black', linewidth=0.5) # Eixo Y
    plt.legend()
    plt.show()

# Chama a função para executar
plot_first_degree_function()
```
